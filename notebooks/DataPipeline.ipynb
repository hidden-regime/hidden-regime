## Data Processing

```python
TICKER = 'AAPL'
```

```python
data = hr.load_stock_data(TICKER, '2023-01-01', '2023-12-31')
```

```python
data.tail(10)
```

```python
fig = plt.figure()
ax = data.log_return.plot()
ax.set_ylabel('Log Return')
ax.set_xlabel('Index')
_ = fig.suptitle(TICKER)
```

```python
date_indexed_data = data.set_index('date')
```

```python
fig = plt.figure()
ax = date_indexed_data.log_return.plot()
ax.set_ylabel('Log Return')
ax.set_xlabel('Date')
_ = fig.suptitle(TICKER)
```
